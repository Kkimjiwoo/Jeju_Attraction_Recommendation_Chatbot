```python
# 각 음식점에 대해 '예약 없이 이용'과 '예약 후 이용'의 총합 계산
df['총합'] = df[['예약 없이 이용', '예약 후 이용']].sum(axis=1)

# 비중 계산
df['예약 없이 이용'] = df['예약 없이 이용'] / df['총합']
df['예약 후 이용'] = df['예약 후 이용'] / df['총합']

# 불필요한 '총합' 컬럼 제거
df.drop(columns=['총합'], inplace=True)
```

```python
# 각 음식점에 대해 대기시간의 빈도수 총합 계산
df['총합'] = df[['바로 입장', '10분 이내','30분 이내', '30분 이상', '1시간 이상', '2시간 이상']].sum(axis=1)

# 비중 계산
df['바로 입장'] = df['바로 입장'] / df['총합']
df['10분 이내'] = df['10분 이내'] / df['총합']
df['30분 이내'] = df['30분 이내'] / df['총합']
df['30분 이상'] = df['30분 이상'] / df['총합']
df['1시간 이상'] = df['1시간 이상'] / df['총합']
df['2시간 이상'] = df['2시간 이상'] / df['총합']

# 불필요한 '총합' 컬럼 제거
df.drop(columns=['총합'], inplace=True)
```

```python
# 각 음식점을 방문하는 목적의 빈도수 총합 계산
df['총합'] = df[['데이트', '친목', '나들이', '여행', '일상', '기념일', '회식', '가족모임', '비지니스']].sum(axis=1)

# 비중 계산
df['데이트'] = df['데이트']/ df['총합']
df['친목'] = df['친목'] / df['총합']
df['나들이'] = df['나들이'] / df['총합']
df['여행'] = df['여행'] / df['총합']
df['일상'] = df['일상'] / df['총합']
df['기념일'] = df['기념일'] / df['총합']
df['회식'] = df['회식'] / df['총합']
df['가족모임'] = df['가족모임'] / df['총합']
df['비지니스'] = df['비지니스'] / df['총합']

# 불필요한 '총합' 컬럼 제거
df.drop(columns=['총합'], inplace=True)
```

```python
# 각 음식점의 동행자 빈도수 총합 계산
df['총합'] = df[['연인・배우자', '친구', '지인・동료', '혼자', '아이', '부모님', '반려동물', '친척・형제', '기타'  ]].sum(axis=1)

# 비중 계산
df['연인・배우자'] = df['연인・배우자']/ df['총합']
df['친구'] = df['친구'] / df['총합']
df['지인・동료'] = df['지인・동료'] / df['총합']
df['혼자'] = df['혼자'] / df['총합']
df['아이'] = df['아이'] / df['총합']
df['부모님'] = df['부모님'] / df['총합']
df['반려동물'] = df['반려동물'] / df['총합']
df['친척・형제'] = df['친척・형제'] / df['총합']
df['기타'] = df['기타'] / df['총합']

# 불필요한 '총합' 컬럼 제거
df.drop(columns=['총합'], inplace=True)
```

```python
from collections import Counter
# 빈도 탐색
counter = Counter(all_pos_tokens)
```

```python
available_counter = Counter({x: counter[x] for x in counter if len(x) > 1})
len(available_counter)
```

```python
#한글자 명사 제거
available_counter = Counter({x: counter[x] for x in counter if len(x) > 1})

# 하위 200개 항목 가져오기
least_common_noun_200 = available_counter.most_common()[-200:-401:-1]  # 빈도수가 가장 낮은 200개

# 키만 추출하기
least_common_noun_200_keys = [item[0] for item in least_common_noun_200]
print(least_common_noun_200_keys)
```

```python
# 상위 200개 항목 가져오기
most_common_noun_200 = available_counter.most_common(500)

# 키만 추출하기
most_common_noun_200_keys = [item[0] for item in most_common_noun_200]

print(most_common_noun_200_keys)
```
