```python
from konlpy.tag import Okt
from collections import Counter
import re
```

```python
# Okt 객체 생성
okt = Okt()

# 명사 추출 함수
def extract_nouns(reviews):
    # 명사 형태소 분석
    nouns = okt.nouns(reviews)
    # 불용어 제외
    filtered_nouns = [noun for noun in nouns if noun not in stopwords]
    return filtered_nouns

# 새로운 컬럼 추가
df['negative_keywords'] = df['review_list_0'].apply(extract_nouns)
df['positive_keywords'] = df['review_list_1'].apply(extract_nouns)
```

```python
classification_keyword = df.copy()
classification_keyword.to_csv('./classification_keyword.csv', encoding='utf-8', index=False)
```
